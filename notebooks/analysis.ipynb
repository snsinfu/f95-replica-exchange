```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
output = np.loadtxt("../output.dat")
```

```python
time = output[:, 0]
x_history = output[:, 1]
e_history = output[:, 2]
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 3))

ax = axes[0]
ax.plot(time, x_history, lw=0.5, color="black")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_yticks([-2, 0, 2])
ax.set_xticklabels([])
ax.set_ylabel("Position")

ax = axes[1]
ax.plot(time, e_history, lw=0.5, color="red")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_xlabel("Time")
ax.set_ylabel("Energy")

pass
```

```python
probs, bin_edges = np.histogram(x_history, bins=31, density=True)
free_energy = -np.log(probs)
coords = (bin_edges[1:] + bin_edges[:-1]) / 2

fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(coords, free_energy, color="royalblue")
ax.set_xlabel("Position")
ax.set_ylabel("Free energy")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

pass
```
